```python
import numpy as np
import pandas as pd

# Helper Functions
def get_envelope_bounds(atr, near_factor, far_factor, envelope):

    upper_far = envelope + far_factor * atr
    upper_near = envelope + near_factor * atr
    lower_far = envelope - far_factor * atr
    lower_near = envelope - near_factor * atr
    upper_avg = ((upper_far + upper_near) / 2)
    lower_avg = ((lower_far + lower_near) / 2)

    return upper_near, upper_far, upper_avg, lower_near, lower_far, lower_avg

def custom_atr(length, high, low, close):
    true_range = []
    true_range.append(np.log(high[0]) - np.log(low[0]))
    for i in range(1, len(high)):
        true_range.append(np.maximum(np.maximum(np.log(high[i]) - np.log(low[i]), np.abs(np.log(high[i]) - np.log(close[i-1]))), np.abs(np.log(low[i]) - np.log(close[i-1]))))
    
    true_range_series = pd.Series(true_range)
    return true_range_series.ewm(span = length, min_periods = length).mean()

def custom_kernel(x, h, alpha, x_0):
    
    sum_weights = 0.0
    sum_x_weights = 0.0
    for y in range(h+1):
        weight = np.power(1 + (np.power((x_0 - y), 2) / (2 * alpha * h * h)), -alpha)
        sum_weights += weight
        sum_x_weights += (weight * x[x_0 - y])
    #return x[x_0]
    return sum_x_weights / sum_weights

# Custom Settings
custom_lookback_window = 8
custom_relative_weighting = 8
#custom_start_regression_bar = 25

def calculate_rsi(close_prices, window=14):
    deltas = np.diff(close_prices)
    seed = deltas[:window+1]
    up = seed[seed >= 0].sum() / window
    down = -seed[seed < 0].sum() / window
    rs = up / down
    rsi = np.zeros_like(close_prices)
    rsi[:window] = 100. - 100. / (1. + rs)

    for i in range(window, len(close_prices)):
        delta = deltas[i - 1]
        if delta > 0:
            upval = delta
            downval = 0.
        else:
            upval = 0.
            downval = -delta

        up = (up * (window - 1) + upval) / window
        down = (down * (window - 1) + downval) / window

        rs = up / down
        rsi[i] = 100. - 100. / (1. + rs)

    return rsi


custom_envelope_close = []
custom_envelope_high = []
custom_envelope_low = []

for i in range (custom_lookback_window, len(prices_day_series)):
    #custom_envelope_close.append(np.exp(custom_kernel(np.log(prices_day_series), custom_lookback_window, custom_relative_weighting, i)))
    custom_envelope_close.append(np.exp(custom_kernel(np.log(prices_day_series), custom_lookback_window, custom_relative_weighting, i)))
    custom_envelope_high.append(np.exp(custom_kernel(np.log(high_day_series), custom_lookback_window, custom_relative_weighting, i)))
    custom_envelope_low.append(np.exp(custom_kernel(np.log(low_day_series), custom_lookback_window, custom_relative_weighting, i)))
# Envelope Calculation
#custom_envelope_close = np.exp(custom_kernel(np.log(prices_series), custom_lookback_window, custom_relative_weighting, custom_start_regression_bar))
#custom_envelope_high = np.exp(custom_kernel(np.log(high_series), custom_lookback_window, custom_relative_weighting, custom_start_regression_bar))
#custom_envelope_low = np.exp(custom_kernel(np.log(low_series), custom_lookback_window, custom_relative_weighting, custom_start_regression_bar))

custom_envelope_close_series = pd.Series(custom_envelope_close)
#custom_envelope_high_series = pd.Series(custom_envelope_high)
#custom_envelope_low_series = pd.Series(custom_envelope_low)


custom_envelope = custom_envelope_close
custom_atr_length = 60
custom_atr_line = custom_atr(custom_atr_length, custom_envelope_high, custom_envelope_low, custom_envelope_close)

custom_near_atr_factor = 2
custom_far_atr_factor = 1.5
print(len(custom_atr_line))
print(len(custom_envelope_close))
custom_upper_near, custom_upper_far, custom_upper_avg, custom_lower_near, custom_lower_far, custom_lower_avg = get_envelope_bounds(custom_atr_line, custom_near_atr_factor, custom_far_atr_factor, np.log(custom_envelope_close))
```
